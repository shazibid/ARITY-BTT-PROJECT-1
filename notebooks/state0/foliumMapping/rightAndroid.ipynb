```python
import pandas as pd
import folium

df = pd.read_csv('../../../data/raw/android_turns.csv')
df = df[(df['eventType'] == 5)]
print(df.shape)
```

```python
df.head()
```

```python
df_highr = df[df.gps_radius>50]
m = folium.Map(location=[df_highr['eventStartLatitude'].mean(), df_highr['eventStartLongitude'].mean()], 
               zoom_start=5,control_scale=True)

# Add points
for _, row in df_highr.iterrows():
    folium.CircleMarker(
        location=[row['eventStartLatitude'], row['eventStartLongitude']],
        radius=4,
        color='blue',
        fill=True,
        fill_opacity=0.7,
        popup=row.get('name', None),
        tooltip = [row['eventStartSpeed'],row['gps_radius'],row['bearingAngleChange']],
        control_scale=True 
    ).add_to(m)

m
```
